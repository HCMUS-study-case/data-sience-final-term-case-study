```python
import numpy as np
import pandas as pd
import datetime
```

### Câu hỏi: Tháng nào trong năm có tổng biến động giá trị dương nhiều nhất, và âm nhiều nhất?
Trả lời được câu hỏi này giúp chúng ta có thể dự đoán quy luật của năm 2023 dựa trên kết quả của năm 2022, để có thể có một cách đầu tư khôn ngoan (tháng nào dương nhiều nhất thì nên đầu tư, và ngược lại)

```python
df_his=pd.read_csv("data-csv/historical_quotes.csv", index_col=0)
df_his['date']=pd.to_datetime(df_his['date'],format='%Y-%m-%d')
df_his.head(10)
```

Lấy dữ liệu từ 1/1/2022 đến ngày hiện tại

```python
first_day = np.datetime64("2022-01-01")
today = np.datetime64('today', 'D')
df_his_ques = df_his[df_his['date'].between(first_day, today)]
df_his_ques.head(10)
```

Hàm hỗ trợ lấy ngày đầu tiên, và ngày cuối cùng của 1 tháng

```python
def getFirstLastDate(month):
    if month==1 or month ==3 or month == 5 or month == 7 or month == 8 or month == 10 or month == 12:
        last = 31
    elif month == 2:
        last = 28
    else:
        last = 30
    
    if month < 10:
        first_day_of_month = np.datetime64("2022-0{}-01".format(month))
        last_day_of_month = np.datetime64("2022-0{}-{}".format(month,last))
    else:
        first_day_of_month = np.datetime64("2022-{}-01".format(month))
        last_day_of_month = np.datetime64("2022-{}-{}".format(month,last))
    tuple_day = (first_day_of_month, last_day_of_month)
    return tuple_day
```

Tìm tháng có tổng biến động giá trị dương nhiều nhất, và âm nhiều nhất?

```python
rate_fluctuation_month = []
rate_fluctuation = df_his_ques["rate fluctuation"]
date_array = df_his_ques["date"]

current_month = 1

while(current_month <= 12):
    tuple_day = getFirstLastDate(current_month)
    sub_df_his_ques = df_his_ques[df_his_ques['date'].between(tuple_day[0], tuple_day[1])]
    rate_sum = sum(sub_df_his_ques["rate fluctuation"])
    rate_fluctuation_month.append(rate_sum)
    current_month+=1
max_month = rate_fluctuation_month.index(max(rate_fluctuation_month)) + 1
min_month = rate_fluctuation_month.index(min(rate_fluctuation_month)) + 1

print("Tháng có sự biến động âm nhiều nhất: ", min_month)
print("Tháng có sự biến động dương nhiều nhất: ", max_month)
```

- Qua kết quả rút ra ở trên, nhóm em nhận thấy trong 2022, tháng 8 là tháng có sự biến động giá cổ phiểu dương nhiều nhất, tức là giá cổ phiếu có xu hướng đi lên nhiều => Nhóm em quyết định đầu tư tất tay vào tháng 8 năm 2023, và sẽ bán trước tháng 10/2023, vì vào tháng 10, giá cổ phiếu có xu hướng giảm.
